```python
def experiment(n):
    birthdays = [randint(1,365) for i in range(n)]
    for i in range(len(birthdays)):
        for j in range(len(birthdays)):
            if birthdays[i] == birthdays[j] and i != j:
                return True
    return False
            
    
```

```python
def experiment(n, bs):
    birthdays = [randint(1,365) for i in range(n)]
    d = {}
    for b in birthdays:
        if b in d:
            d[b] = d[b] + 1
            if d[b] == bs:
                return True
        else:
            d[b] = 1
    return False
```

```python
experiment(100, 4)
```
